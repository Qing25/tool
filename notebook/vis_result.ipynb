```python
from qdls.data import load_json

# data = load_json("/home/qing/raid/paperwork/kgtool/src_pure/results/sampled_50_gpt-4o-mini_v1.json")
data = load_json("/home/qing/raid/paperwork/kgtool/src_pure/results/sampled_50_gpt-4o.json")
```

```python
def vis(sample):
    print(sample["question"])
    print(sample["answer"])
    print(sample["kopl"])
    print("="*100)
    
    for i, step in enumerate(sample['history']):
        print("#"*50 + f" Step {i+1} " + "#"*50)
        print("-"*50 + " Thought " + "-"*50)
        print(step['thought'])
        print("-"*50 + " Action " + "-"*50)
        print(step['action'])
        print("-"*50 + " Result " + "-"*50)
        print(step['result'])
    
    print("$"*50 + " Final Answer " + "$"*50)
    print(sample['response'])


for s in data:
    if len(s['history']) > 10:
        vis(s)
        break

```
